# Scikit Toy datasets

```python
import sklearn.datasets as sk
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# function to save toy datasets to csv files

def saveData(x, y, name, outputDir = "data"):
    if x.shape[1] == 3:
        df = pd.DataFrame(x, columns=["x", "y", "z"])
        df["target"] = y
    elif x.shape[1] == 2:
        df = pd.DataFrame(x, columns=["x", "y"])
        df["target"] = y

    os.makedirs(outputDir, exist_ok=True)
    path = os.path.join(outputDir, f"{name}.csv")
    df.to_csv(path, index=False)
```

### Swiss roll dataset (1000 samples, varied Gaussian noise)

```python
def plotSwiss(X, t):
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')

    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=t, cmap=plt.cm.Spectral, s=50, alpha=0.8)

    ax.set_title("Swiss Roll Toy Dataset")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z ")
    ax.view_init(elev=10., azim=-66)
    plt.show()
```

```python
# swiss roll - 1000 samples, 0.05 gaussian noise

swiss1data, swiss1target = sk.make_swiss_roll(n_samples=1000, noise=0.05)
saveData(swiss1data, swiss1target, "swiss_0.05noise")
```

```python
plotSwiss(swiss1data, swiss1target)
```

```python
# swiss roll - 1000 samples, 0.5 gaussian noise

swiss2data, swiss2target = sk.make_swiss_roll(n_samples=1000, noise=0.5)
plotSwiss(swiss2data, swiss2target)

saveData(swiss2data, swiss2target, "swiss_0.5noise")
```

```python
# swiss roll - 1000 samples, 1.0 gaussian noise

swiss3data, swiss3target = sk.make_swiss_roll(n_samples=1000, noise=1)
plotSwiss(swiss3data, swiss3target)

saveData(swiss3data, swiss3target, "swiss_1noise")
```

### Circle Toy Datasets - 2D concentric circles (1000 samples, varied Gaussian noise)

```python
def plotCircle(X, Y):
    df = pd.DataFrame(dict(x=X[:, 0], y=X[:, 1], label=Y))

    colors = {0: "red", 1: "blue"}
    fig, ax = plt.subplots(figsize=(6, 6))
    grouped = df.groupby("label")
    for key, group in grouped:
        group.plot(ax=ax, kind="scatter", x="x", y="y", label=f"Class {key}", color=colors[key])

    plt.title("Circles Toy Dataset")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend()
    plt.show()
```

```python
# circles - 1000 samples, 0.05 gaussian noise

circle1X, circle1Y = sk.make_circles(n_samples = 1000, noise = 0.05)
plotCircle(circle1X, circle1Y)

saveData(circle1X, circle1Y, "circle_0.05noise")
```

```python
# circles - 1000 samples, 0.2 gaussian noise

circle2X, circle2Y = sk.make_circles(n_samples = 1000, noise = 0.2)
plotCircle(circle2X, circle2Y)

saveData(circle2X, circle2Y, "circle_0.2noise")
```

```python
# circles - 1000 samples, 0.5 gaussian noise

circle3X, circle3Y = sk.make_circles(n_samples = 1000, noise = 0.5)
plotCircle(circle3X, circle3Y)

saveData(circle3X, circle3Y, "circle_0.5noise")
```
